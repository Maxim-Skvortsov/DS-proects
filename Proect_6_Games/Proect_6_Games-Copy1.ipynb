### 'year_of_release','critic_score'

## Проверка гипотез

###  Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

H0 = Нулевая гипотеза - что средний рейтинги пользователей по плотформе Xbox не отличается от рейтинга пользователей по плотформе PC.

H1 = Альтернативная гипотеза - то средний рейтинги пользователей по плотформе  Xbox  отличается от рейтинга пользователей по плотформе PC.


```python
df[df['platform'] == 'pc']['user_score'].hist(alpha=0.7,color='r',figsize=(12, 6), label="PC")
df[df['platform'] == 'xone']['user_score'].hist(alpha=0.5,color='b',figsize=(12, 6), label='Xbox One')
plt.title('Распределение user_score по платформе PC и Xbox One')
plt.ylabel('Колличество')
plt.xlabel('user_score')
plt.legend()
plt.show()
```

У нас есть 2 выборки поэтому будем использовать гипотезу о равенстве средних двух генеральных совокупностей. Применим ttest.

```python
results = st.ttest_ind(
    df[df['platform'] == 'pc']['user_score'].dropna(),
    df[df['platform'] == 'xone']['user_score'].dropna(),
    equal_var=False)
alpha = .05 #критический уровень статистической значимости
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

H0 = Нулевая гипотеза - средний рейтинги пользователей по жанру Action не отличается от рейтинга пользователей по жанру Sports.

H1 = Альтернативная гипотеза - средний рейтинги пользователей по жанру Actionx  отличается от рейтинга пользователей жанру Sports.

```python
df[df['genre'] == 'action']['user_score'].hist(alpha=0.7,color='r',figsize=(12, 6), label="Action")
df[df['genre'] == 'sports']['user_score'].hist(alpha=0.5,color='b',figsize=(12, 6), label='Sports')
plt.title('Распределение user_score по жанру Action и Sports')
plt.ylabel('Колличество')
plt.xlabel('user_score')
plt.legend()
plt.show()
```

У нас есть 2 выборки поэтому будем использовать гипотезу о равенстве средних двух генеральных совокупностей. Применим ttest.

```python
results = st.ttest_ind(
    df[df['genre'] == 'action']['user_score'].dropna(),
    df[df['genre'] == 'sports']['user_score'].dropna(),
    equal_var=False)
alpha = .05 # критический уровень статистической значимости
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод

Провели ttest по Гипотеза о равенстве средних двух генеральных совокупностей:

- Предположительно Нулевая гипотеза - что средний рейтинги пользователей по плотформе Xbox не отличается от рейтинга пользователей по плотформе PC оаказалась верной. 
- Предположительно Альтернативная гипотеза - средний рейтинги пользователей по жанру Actionx отличается от рейтинга пользователей жанру Sports оказалась верной.
