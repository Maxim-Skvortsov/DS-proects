# Определение перспективного тарифа для телеком-компании

У вас есть данные компании «Мегалайн» — федерального оператора сотовой связи. Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.
Вам предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов. В вашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

### Объединение таблиц

Для облегчения поиска и работы с данными объединим таблицы в одну.

calls internet и messages  сгрупперуем каждую таблицу по месяцам и объеденим поочередно  по столбцам 'user_id' и 'month' в таблице df. Далее по 'user_id' объединим с  users. И по 'tarif' объеденим с  tariffs

df_calls сколько по времени и по колличеству звонков в месяц сделал пользователь

```python
df_calls = df_calls.groupby(['user_id', 'month']).agg({'duration': ['sum','count']})
```

```python
df_calls.columns = ["_".join(a) for a in df_calls.columns.to_flat_index()]
```

```python
df_calls = df_calls.reset_index()
```

```python
df_calls.head()
```

df_internet сколько мб в месяц потратил. Компания считает в гигабайтах за месяц и  округляет в бо́льшую сторону Поэтому переведем мегабайты в гигабайты и округлим вверх.

```python
df_internet = df_internet.groupby(['user_id', 'month']).agg({'mb_used': ['sum']})
```

```python
df_internet.columns = ["_".join(a) for a in df_internet.columns.to_flat_index()]
```

```python
df_internet = df_internet.reset_index()
```

```python
df_internet['mb_used_sum'] /= 1024
df_internet['mb_used_sum'] = np.ceil(df_internet['mb_used_sum']).astype('int')
```

```python
df_internet
```

```python
df_messages.columns = ["_".join(a) for a in df_messages.columns.to_flat_index()]
```

```python
df_messages = df_messages.reset_index()
```

**Тариф «Смарт»**

Ежемесячная плата: 550 рублей.Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика.

Стоимость услуг сверх тарифного пакета:

- сообщение: 3 рубля
- 1 Гб интернет-трафика: 200 рублей.

**Тариф «Ультра»**

Ежемесячная плата: 1950 рублей. Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
Стоимость услуг сверх тарифного пакета:
- минута разговора: 1 рубль; 
- сообщение: 1 рубль; 
- 1 Гб интернет-трафика: 150 рублей.

Исходя из этого расчитаем сколько ползователь заплатил по тарифу в этом месяце

```python
def monthly_revenue(row):
    if row['calls_duration'] > row['minutes_included']:
        minuts_is = (row['calls_duration']-row['minutes_included'])*row['rub_per_minute']
    else:
        minuts_is = 0
    if row['message_count'] > row['messages_included']:
        messages_is = (row['message_count']-row['messages_included'])*row['rub_per_message'] 
    else:
         messages_is = 0
    if row['gb_used'] > row['gb_per_month_included']:
        internet_is = (row['gb_used']-row['gb_per_month_included']) * row['rub_per_gb']
    else:
         internet_is = 0
    return int(minuts_is + messages_is + internet_is + row['rub_monthly_fee'])
```

```python
df['payment']= df.apply(monthly_revenue,axis=1)
```

## Проверка гипотез

средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.
Пороговое значение alpha задайте самостоятельно.
Поясните:

### Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

H0 Нулевая гипотеза что средняя выручка пользователей тарифов Smart и Ultra одинакова
 H1 Альтернативная гипотеза что средняя выручка пользователей тарифов Smart и Ultra различаются

```python
gip_smart = df[df['tariff'] == 'smart']['payment']
gip_ultra = df[df['tariff'] == 'ultra']['payment']
```

```python
gip_ultra.hist(alpha=0.7,color='r',figsize=(12, 6), label="Ultra")
gip_smart.hist(alpha=0.7,color='skyblue',figsize=(12, 6), label='Smart')
plt.title('Распределение payment')
plt.ylabel('Колличество')
plt.xlabel('payment')
plt.legend()
plt.show()
```

У нас есть 2 выборки поэтому будем использовать гипотезу о равенстве средних двух генеральных совокупностей

```python
results = st.ttest_ind(gip_smart, gip_ultra, equal_var=False)

alpha = .05 # критический уровень статистической значимости
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

H0 =  Нулевая гипотеза -  что средняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов.

H1 =  Альтернативная гипотеза - что средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.
Гипотеза о равенстве средних для зависимых (парных) выборок

```python
gip_mosc = df[df['city'] == 'москва']['payment']
gip_etc = df[df['city'] != 'москва']['payment']
```

```python
gip_mosc.hist(alpha=0.7,color='r',figsize=(12, 6), label="Москва")
gip_etc.hist(alpha=0.7,color='skyblue',figsize=(12, 6), label='Остальные города')
plt.title('Распределение payment')
plt.ylabel('Колличество')
plt.xlabel('payment')
plt.legend()
plt.show()
```

У нас есть 2 выборки поэтому будем использовать гипотезу о равенстве средних двух генеральных совокупностей

```python
results = st.ttest_ind(gip_mosc, gip_etc)
alpha = .05 # критический уровень статистической значимости
print('p-значение: ', results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод

Провели ttest по Гипотеза о равенстве средних двух генеральных совокупностей:
    
- Предположительно нулевая гипотеза - что средняя выручка пользователей тарифов Smart и Ultra одинакова оказалась неверной. 
- Предположительно нулевая гипотеза - что средняя выручка пользователей из Москвы не отличается от выручки пользователей из других регионов оказалась верная

А значит выручка Smart и Ultra предположительно различаются. А Москвы и других регионов предположительно не различаются.
